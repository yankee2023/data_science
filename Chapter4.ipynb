```python
# Goal
# 確立と統計の基礎的な理解と計算ができる

# ライブラリのimport
import numpy as np
import scipy as sp
import pandas as pd
from pandas import Series, DataFrame

import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
%matplotlib inline

# 小数第三位まで表示
%precision 3

# ランダムシード固定
np.random.seed(0)
```

```python
# 数学的確率

# さいころがとりうる値を配列に格納
dice_data = np.array([1, 2, 3, 4, 5, 6])
```

```python
print('1つだけランダムに抽出: ', np.random.choice(dice_data, 1))
```

```python
# 1つずつの試行結果を根元事象（基本事象）
# ex) 5
# すべての可能な根元事象を集めた集合を標本空間
# S = { 1, 2, 3, 4, 5, 6}
# 標本空間の任意の部分集合を事象
# ex) 5が出る事象X
```

```python
# 統計的確率
# サイコロを1000回振るシミュレーション

calc_step = 1000

dice_rolls = np.random.choice(dice_data, calc_step)

# 各数字がどれくらいの割合で抽出された計算
for i in range(1, 7):
    p = len(dice_rolls[dice_rolls == i]) / calc_step
    print(i, 'が出る確率: ', p)
```

```python
# 練習問題4-1
coin_data = np.array([0,1])
coin_tos = np.random.choice(coin_data, calc_step)
for i in range(0, 2):
    p = len(coin_tos[coin_tos == i]) / calc_step
    if (i == 0):
        print('表が出る確率: ', p)
    else:
        print('裏が出る確率: ', p)
```
