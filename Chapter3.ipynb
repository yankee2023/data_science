```python
# 単回帰分析
## アウトプット(目的変数)とインプット(説明変数)の関係で
## インプットが1変数のものを単回帰分析、2変数以上あるものを重回帰分析という。
plt.plot(student_data_math['G1'], student_data_math['G3'], 'o')
plt.xlabel('G1 grade')
plt.ylabel('G3 grade')
plt.grid(True)
```

```python
# 線形回帰分析

from sklearn import linear_model

# 線形回帰のインスタンス生成
reg = linear_model.LinearRegression()
```

```python
# 説明変数:X
# 目的変数:Y
X = student_data_math.loc[:, ['G1']].values
Y = student_data_math['G3'].values

# 予測モデルの計算
reg.fit(X, Y)

# 回帰係数
print('回帰係数: ', reg.coef_)

# 切片
print('切片: ', reg.intercept_)
```

```python
# y=ax+b
# 回帰式: y = 1.106x+(-1.65....)を描く

plt.scatter(X, Y)
plt.xlabel('G1 grade')
plt.ylabel('G3 grade')

plt.plot(X, reg.predict(X))
plt.grid(True)
```

```python
# 決定係数（寄与率）の計算
# 1に近いほど良いモデルとなる。
print('決定係数: ', reg.score(X, Y))
```

```python
# 練習問題3-4
student_por = pd.read_csv('student-por.csv', sep=';')

X_por = student_por.loc[:, ['G1']].values
Y_por = student_por['G3'].values
reg.fit(X_por, Y_por)

print('回帰係数: ', reg.coef_)
print('切片: ', reg.intercept_)
print('決定係数: ', reg.score(X_por, Y_por))
```

```python
# 練習問題3-5
plt.scatter(X_por, Y_por)
plt.xlabel('G1 grade')
plt.ylabel('G3 grade')
plt.plot(X_por, reg.predict(X_por))
plt.grid(True)
```

```python
# 練習問題3-6
X_por = student_por.loc[:, ['absences']].values
Y_por = student_por['G3'].values
reg.fit(X_por, Y_por)

print('回帰係数: ', reg.coef_)
print('切片: ', reg.intercept_)
print('決定係数: ', reg.score(X_por, Y_por))

plt.scatter(X_por, Y_por)
plt.xlabel('absences')
plt.ylabel('G3 grade')
plt.plot(X_por, reg.predict(X_por))
plt.grid(True)
```
